```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from google.colab import drive

# ----------------------------
# 0. Mount Google Drive
# ----------------------------
drive.mount('/content/drive')

# ----------------------------
# 1. Load your trained YOLOv8 model
# ----------------------------
best_model_path = '/content/drive/MyDrive/Colab Notebooks/BrainTumor/BrainTumorYolov8/runs/train/weights/best.pt'
best_model = YOLO(best_model_path)

# ----------------------------
# 2. Normalization and resizing functions
# ----------------------------
def normalize_image(image):
    return image / 255.0

def resize_image(image, size=(640, 640)):
    return cv2.resize(image, size)

# ----------------------------
# 3. Load validation/test images from Google Drive
# ----------------------------
valid_images_path = '/content/drive/MyDrive/Colab Notebooks/BrainTumor/BrainTumorYolov8/test/images'  # Update your Drive path here

# Get list of all jpg/png images
image_files = [file for file in os.listdir(valid_images_path) if file.endswith(('.jpg', '.png'))]

if len(image_files) == 0:
    raise Exception("No images found in the test directory!")

# Select 16 images at equal intervals for 4x4 grid
num_images = len(image_files)
step_size = max(1, num_images // 16)
selected_images = [image_files[i] for i in range(0, num_images, step_size)]

# ----------------------------
# 4. Plot predictions
# ----------------------------
fig, axes = plt.subplots(4, 4, figsize=(20, 21))
fig.suptitle('TEST Set Inferences', fontsize=24)

for i, ax in enumerate(axes.flatten()):
    if i < len(selected_images):
        image_path = os.path.join(valid_images_path, selected_images[i])
        image = cv2.imread(image_path)

        if image is not None:
            # Resize & normalize
            resized_image = resize_image(image)
            normalized_image = normalize_image(resized_image)
            normalized_image_uint8 = (normalized_image * 255).astype(np.uint8)

            # YOLOv8 prediction
            results = best_model.predict(source=normalized_image_uint8, imgsz=640, conf=0.5)

            # Annotate & convert color for plotting
            annotated_image = results[0].plot(line_width=2)
            annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
            ax.imshow(annotated_image_rgb)
        else:
            print(f"Failed to load image {image_path}")

    ax.axis('off')

plt.tight_layout()
plt.show()
```
